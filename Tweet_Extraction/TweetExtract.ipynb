```
import tweepy
import csv
import pandas as pd
```

```
consumer_key = 'YOURKEY
consumer_secret = 'YOURKEY'
access_token = 'YOURTOKEN'
access_token_secret = 'YOURTOKEN'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)

csvFile = open("/content/drive/My Drive/tweets.csv","w")
fieldname = ["TweetCreated","UserID","Text","Score"]
w = csv.DictWriter(csvFile, fieldnames = fieldname)
w.writeheader()


for tweet in tweepy.Cursor(api.search,q="#indigoairlines",count=5000,lang="en",since="2020-01-01",tweet_mode='extended').items():
  if (not tweet.retweeted) and ('RT @' not in tweet.full_text):
    print (tweet.created_at,tweet.user.screen_name, tweet.full_text, "\n\n")
    w.writerow({"TweetCreated":tweet.created_at,"UserID":tweet.user.screen_name,"Text":tweet.full_text})

csvFile.close()
```
